```python
# Missing Values

#checking for missing values
df.isnull()

df.notnull()

# checking for particulae column for missing values
chkAge = pd.isnull(df['Age'])
df[chkAge]


chkAge1 = pd.notnull(df['Age'])
df[chkAge1]
```

```python
# Filling the missing values

df.fillna(0) # fills with 0

df.fillna(method = 'pad')# fills null values with previous ones

df.fillna(method = 'bfill')# fills null values with next ones

df['Age'].fillna("No Age",inplace = True)# fills null values with particular data
#df

df.replace(to_replace = np.nan,value = -100000000000000)# repalces null values with data

df.interpolate(method = "linear" ,limit_direction = 'Forward')# fills null values in frwd manner

df.dropna()#drop Rows/Columns of datasets with Null values
```
