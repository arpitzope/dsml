```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
sns.boxplot(data=df,y='math score',x='test preparation course')
```

```python
df
```

```python
#plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='math score',x='parental level of education')
```

```python
#plt.figure(figsize=(10,10),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education',hue='test preparation course',split=True,inner='stick',bw=0.5)
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)
plt.legend(bbox_to_anchor=(1.35,0.5))
```

```python
sns.boxenplot(x='math score',y='test preparation course',data=df,hue='gender')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',shade=True,hue='gender')
```

```python
sns.pairplot(data=df,hue='gender',diag_kind='hist')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch',row='test preparation course')
```

```python
g=sns.PairGrid(df,hue='gender')

g=g.map_upper(sns.scatterplot)

g=g.map_diag(sns.kdeplot)
g=g.map_lower(sns.kdeplot)
```
